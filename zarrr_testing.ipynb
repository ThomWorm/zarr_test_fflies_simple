```python
import xarray as xr
import zarr
```

```python
# Open multiple NetCDF files from the "data" folder and combine them - change this to your folder 
ds = xr.open_mfdataset("data/*.nc", combine="by_coords")

# Save the combined dataset as a Zarr data store
ds.to_zarr("simple_dataset.zarr", mode="w")
```

### serve the zarr dataset

command line: python -m http.server --directory simple_dataset.zarr/ --bind 127.0.0.1 8000

```python
served_data = xr.open_zarr("http://127.0.0.1:8000/")
```

```python
served_data["tmax"].isel(t=0).plot()  #confirm it works with a plot
```
